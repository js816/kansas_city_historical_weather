```python
# Loading the dependencies
import pandas as pd
```

```python
# Reading the monthly mean temp
monthly_mean_temp_df = pd.read_csv('Resources/clean_kc_historic_monthly_mean_temp.csv')
monthly_mean_temp_df
```

```python
# Looking for null values
monthly_mean_temp_df.isnull().sum()
```

```python
# Looking at DataTypes
monthly_mean_temp_df.dtypes
```

```python
# Looking at summary statistics
monthly_mean_temp_df.describe()
```

```python
# Looking at January series
monthly_mean_temp_df['January']
```

```python
# Changing shape of data
# Adapting code found here:  https://www.datasciencemadesimple.com/reshape-wide-long-pandas-python-melt-function/
monthly_mean_temp_long_df = pd.melt(monthly_mean_temp_df, id_vars='Year', var_name='month', value_name='average temp')
monthly_mean_temp_long_df
```

```python
# Exporting long mean temp as CSV
monthly_mean_temp_long_df.to_csv('Resources/long_monthly_mean_temp.csv', index=False)
```

```python
# Reading in 1889 and 2021 data
december_df = pd.read_csv('Resources/december_1889_and_2021.csv')
december_df
```
